# Smeaton 300

```python
import os
from events import (
    load_raw_data, filter_signature_events, filter_by_project, filter_public_events, filter_school_events, prepare, save_files, count_wards
)

import ticket_helpers
```

Set output directory

```python
OUT_DIR = '../../../../docs/metrics/events/smeaton-300/_data/events'
```

Read in event data and filter by **Smeaton 300** projects.

The criteria are:

* project type is 'Produce (Signature)' (in the load_event_data function)

It's then filtered by project_name is '10 - Melt'

```python
project_names = [
  '09 - Smeaton300: Moon Palace East Leeds Observatory',
  '09 - Smeaton300: Learning programme',
  '09 - Smeaton300: Talks and Discussions'
]
data = (
  load_raw_data()
    .pipe(filter_signature_events)
)
data = data.loc[data.project_name.isin(project_names)]
```

Run the processing and save the files

```python
data.pipe(prepare).pipe(save_files, OUT_DIR)
```

Filter by Moon Palace events

```python
MOON_PALACE_EVENTS_DIR = OUT_DIR+'/moon_palace'
os.makedirs(MOON_PALACE_EVENTS_DIR, exist_ok=True)


data.pipe(filter_by_project, '09 - Smeaton300: Moon Palace East Leeds Observatory').pipe(prepare).pipe(save_files, MOON_PALACE_EVENTS_DIR)
```

Filter by public events

```python
PUBLIC_EVENTS_DIR = OUT_DIR+'/public'
os.makedirs(PUBLIC_EVENTS_DIR, exist_ok=True)


data.pipe(filter_public_events).pipe(prepare).pipe(save_files, PUBLIC_EVENTS_DIR)
```

Filter by Schools events

```python
SCHOOLS_EVENTS_DIR = OUT_DIR+'/schools'
os.makedirs(SCHOOLS_EVENTS_DIR, exist_ok=True)


data.pipe(filter_school_events).pipe(prepare).pipe(save_files, SCHOOLS_EVENTS_DIR)
```

Ticket Processing

```python
ticket_sales = data.pipe(filter_public_events).pipe(ticket_helpers.get_tickets_for_event)
```

By ward

```python
ticket_sales.pipe(ticket_helpers.summarise_by_ward).reset_index().to_csv(
    os.path.join(OUT_DIR, 'tickets_by_ward.csv'), index=False)
```

By date

```python
(
  ticket_sales
    .pipe(ticket_helpers.summarise_by_date)
    .resample('W-FRI').sum()
    .cumsum().to_csv(os.path.join(OUT_DIR, 'tickets_by_date.csv'))
)
```
